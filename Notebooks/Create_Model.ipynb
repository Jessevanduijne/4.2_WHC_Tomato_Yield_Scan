```python
# Divide in to training and testing sets
from os import makedirs
from os import listdir
from shutil import copyfile
from random import seed
from random import random

# dataset directory
pwd = '../Dataset/'
dataset_home = pwd + 'healthy_vs_unhealthy_tomatoes/'

# Create sub-directories as train and test. This test is actually the validation dataset so you should have a separate folder, for your convenience you can rename it as validation too
subdirs = ['train/', 'test/']
for subdir in subdirs:
    # create label subdirs
    labeldirs = ['healthy/', 'unhealthy/']
    for labldir in labeldirs:
            newdir = dataset_home + subdir + labldir
            makedirs(newdir, exist_ok=True)
```

```python
# used to save the state of random function, so that it can generate some random numbers on multiple execution of the code on the same machine or on different machines 
seed(1)

# define ratio of pictures to be used for validation and range varies from 0.2 to 0.3
validation_ratio = 0.2

# copy training dataset and validation dataset into respective directories,
healthy_directory = pwd + 'Good v1 (total)/'
unhealthy_directory = pwd + 'Bad v1 (total)/'

# divide datasets for healthy tomatoes
for file in listdir(healthy_directory):
    src = healthy_directory + file
    dst_dir = 'train/healthy/'
    if random() < validation_ratio:
        dst_dir = 'test/healthy/'
        dst = dataset_home + dst_dir + file
        copyfile(src, dst)
    else:
        dst = dataset_home + dst_dir + file
        copyfile(src, dst)
        
# divide datasets for unhealthy tomatoes
for file in listdir(unhealthy_directory):
    src = unhealthy_directory + file
    dst_dir = 'train/unhealthy/'
    if random() < validation_ratio:
        dst_dir = 'test/unhealthy/'
        dst = dataset_home + dst_dir + file
        copyfile(src, dst)
    else:
        dst = dataset_home + dst_dir + file
        copyfile(src, dst)
```

## CNN has following layers

- Convolution Layer
- ReLU Layer
- Pooling Layer
- Fully Connected Layer

```python
# importing libraries
from keras.models import Sequential
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, Flatten, MaxPooling2D, Dropout, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
import tensorflow as tf

input_width = 300
input_height = 300

# Defining model 
model = tf.keras.Sequential()

# Block 1
model.add(Conv2D(32, (3, 3), activation='relu', kernel_initializer='he_uniform', padding='same', input_shape=(input_width, input_height, 3)))
model.add(MaxPooling2D((2, 2)))

model.add(Flatten())
model.add(Dense(128, activation='relu', kernel_initializer='he_uniform'))
model.add(Dense(1, activation='sigmoid'))

# compile model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# we need to rescale the pixel values as our original images are in RGB with coefficients in the 0-255, but such values would be too high for our model to process
datagen = ImageDataGenerator(rescale=1.0/255.0)

# prepare Train and Validation dataset
trainDS = datagen.flow_from_directory(dataset_home + 'train/',
class_mode='binary', batch_size=64, target_size=(input_width, input_height))
testDS= datagen.flow_from_directory(dataset_home + 'test/',
class_mode='binary', batch_size=64, target_size=(input_width, input_height))

# Fitting the model, this will take some time as all epochs will be completed. Once done, the history carries four list values i.e. loss, accuracy, val_accuracy, and val_loss.
history = model.fit_generator(trainDS, steps_per_epoch=len(trainDS),
validation_data=testDS, validation_steps=len(testDS), epochs=20, verbose=0)

# evaluate model
_, accuracy = model.evaluate_generator(testDS, steps=len(testDS), verbose=0)
print('> %.3f' % (accuracy * 100.0))
```

```python
pyplot.style.use('fivethirtyeight')

pyplot.plot(history.history['accuracy'])
pyplot.plot(history.history['val_accuracy'])
pyplot.title('Model accuracy')
pyplot.ylabel('Accuracy')
pyplot.xlabel('Epoch')
pyplot.legend(['Train', 'Val'], loc='upper left')
pyplot.show()
```

```python
pyplot.plot(history.history['loss'])
pyplot.plot(history.history['val_loss'])
pyplot.title('Model loss')
pyplot.ylabel('Loss')
pyplot.xlabel('Epoch')
pyplot.legend(['Train', 'Val'], loc='upper right')
pyplot.show()
```
